## Initialise your pipleine with the model you just fine-tuned

Note: The model name should be in the format: "username/model_name", for the sake of simplicity in the demo I am going to just use the vanilla pre-trained Wav2Vec2 model.

```python
from transformers import pipeline

p = pipeline("automatic-speech-recognition", "facebook/wav2vec2-base-960h")
```

## Define the Gradio interface for ASR

Note: That you can also host this on Hugging Face spaces with the same code: https://hf.co/spaces

```python
import gradio as gr

def transcribe(audio):
    text = p(audio)["text"]
    return text

gr.Interface(
    fn=transcribe, 
    inputs=gr.Audio(source="microphone", type="filepath"), 
    outputs="text").launch()
```
